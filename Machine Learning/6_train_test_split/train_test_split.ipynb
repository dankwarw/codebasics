<p><b>We have a dataset containing prices of used BMW cars. We are going to analyze this dataset
and build a prediction function that can predict a price by taking mileage and age of the car
as input. We will use sklearn train_test_split method to split training and testing dataset</b></p>

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python


df = pd.read_csv("carprices.csv")
df.head()
```

```python
plt.scatter(df.Mileage,df["Sell Price($)"])
plt.title("Scatter plot")
plt.xlabel("Mileage")
plt.ylabel("Price")
plt.show()
```

```python
plt.scatter(df["Age(yrs)"],df["Sell Price($)"])
plt.title("Scatter plot")
plt.xlabel("Mileage")
plt.ylabel("Age(yrs)")
plt.show()
```

```python
X = df[["Age(yrs)","Mileage"]]
y = df["Sell Price($)"]
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=12)
```

```python
len(X_test) # 20percent of your data
```

```python
X_train  #because of random_state, we do not see a change in the data
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
y_test
```

```python
model.score(X_test,y_test)
```
